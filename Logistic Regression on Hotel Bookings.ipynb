# outliers

```python
df.boxplot(column=["lead_time"])
plt.show()
```

```python
def remove_outlier(col):
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
low_leadtime,upper_leadtime=remove_outlier(df["lead_time"])
df["lead_time"]=np.where(df["lead_time"]>upper_leadtime,upper_leadtime,df["lead_time"])
df["lead_time"]=np.where(df["lead_time"]<low_leadtime,low_leadtime,df["lead_time"])
```
